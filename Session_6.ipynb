## **Order Calculator**

```python
def calculate_order(price, quantity, tax=0.05, discount=0, shipping=100):
    """price: product price
        quantity: number of items
        tax: tax percentage (default 5%)
        discount: discount amount (default 0)
        shipping: shipping charges (default 100)"""

    subtotal = price * quantity
    tax_amount = subtotal * tax
    total = subtotal + tax_amount + shipping - discount
    print(subtotal)
    print(tax_amount)


    return total
```

```python
calculate_order(5000,quantity=5,tax=0.02)
```

## **Employee Salary Calculator**

```python
def calculate_salary(basic_salary, bonus=0, overtime_hours=0, tax_rate=0.1):
    """
    basic_salary: employee base salary
    bonus: extra bonus amount
    overtime_hours: extra working hours
    tax_rate: tax percentage (default 10%)
    """

    overtime_pay = overtime_hours * 500   # 500 per hour
    gross_salary = basic_salary + bonus + overtime_pay
    tax_amount = gross_salary * tax_rate
    net_salary = gross_salary - tax_amount


    return net_salary
```

```python
calculate_salary(100000)
```
